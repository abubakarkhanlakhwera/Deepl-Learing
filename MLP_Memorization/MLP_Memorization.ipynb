**Before Memorization**

```python
# fibanacci
import time
def fib(n):
  if n==1 or n==0:
    return 1
  else:
    return fib(n-1)+fib(n-2)
```

```python
start = time.time()
fib(40)
end = time.time()
print(end-start)
```

**After memorization**

```python
def fib(n,d):
  if n in d:
    return d[n]
  else:
    d[n] =  fib(n-1,d)+fib(n-2,d)
    return d[n]
```

```python
start = time.time()
d= {0:1,1:1}
fib(40,d)
end = time.time()
print(end-start)
```
