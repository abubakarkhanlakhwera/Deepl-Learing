```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/placement.csv')
```

```python
print(df.shape)
print(df.head())
```

```python
sns.scatterplot(x='cgpa',y='resume_score',hue=df['placed'],data=df)
```

```python
X= df.iloc[:,:-1]
y= df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values,y.values,clf=p,legend=2)
```
