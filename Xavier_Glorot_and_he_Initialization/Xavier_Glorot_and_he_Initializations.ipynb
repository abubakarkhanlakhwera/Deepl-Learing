# Xavier (Glorot) and He Initialization

Both Xavier (Glorot) and He Initialization are widely used techniques for weight initialization in neural networks. They address the issues of exploding and vanishing gradients by ensuring proper variance of the weights at the start of training.

---

## 1. **Xavier Initialization (Glorot Initialization)**

### Overview:
- Proposed by Glorot and Bengio (2010).
- Designed for activation functions like **sigmoid** or **tanh**.
- Balances the variance of activations and gradients across layers.

### Formula:
For a layer with `n_in` input neurons and `n_out` output neurons:
- **Uniform Distribution:**  
  \( W \sim \mathcal{U}\left(-\frac{\sqrt{6}}{\sqrt{n_{\text{in}} + n_{\text{out}}}}, \frac{\sqrt{6}}{\sqrt{n_{\text{in}} + n_{\text{out}}}}\right) \)
- **Normal Distribution:**  
  \( W \sim \mathcal{N}\left(0, \frac{1}{n_{\text{in}} + n_{\text{out}}}\right) \)

### Key Properties:
- Prevents vanishing/exploding gradients for activation functions with outputs near zero.
- Suitable for shallower networks or those using saturating activations (e.g., sigmoid, tanh).

---

## 2. **He Initialization**

### Overview:
- Proposed by He et al. (2015).
- Optimized for **ReLU** and its variants (e.g., Leaky ReLU, ELU).
- Adjusts variance based on the number of input neurons, accommodating the non-linear behavior of ReLU.

### Formula:
For a layer with `n_in` input neurons:
- **Uniform Distribution:**  
  \( W \sim \mathcal{U}\left(-\sqrt{\frac{6}{n_{\text{in}}}}, \sqrt{\frac{6}{n_{\text{in}}}}\right) \)
- **Normal Distribution:**  
  \( W \sim \mathcal{N}\left(0, \frac{2}{n_{\text{in}}}\right) \)

### Key Properties:
- Ensures gradients are well-scaled through the network when using ReLU.
- Handles sparsity introduced by ReLU (zeros for negative inputs).

---

## Comparison: Xavier vs. He Initialization

| **Aspect**               | **Xavier Initialization**                  | **He Initialization**                 |
|--------------------------|---------------------------------------------|----------------------------------------|
| **Use Case**             | Sigmoid, Tanh                              | ReLU, Leaky ReLU, ELU                 |
| **Variance Scaling**     | Based on \( 1/(n_{\text{in}} + n_{\text{out}}) \) | Based on \( 2/n_{\text{in}} \)        |
| **Prevents**             | Vanishing gradients                        | Exploding gradients                   |
| **Distribution**         | Uniform or Normal                          | Uniform or Normal                     |
